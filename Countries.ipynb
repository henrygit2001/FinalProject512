```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
data = pd.read_csv('world-data-2023.csv').dropna(axis=0)
data.head()
```

```python
print(data.shape)
```

```python
brush = alt.selection_interval(
    encodings=['x'] # limit selection to x-axis (year) values
)

# dynamic query histogram
birth_rate = alt.Chart(data).mark_bar().add_params(
    brush
).encode(
    alt.X('Birth Rate:Q', title='Films by Release Year'),
    alt.Y('count():Q', title=None)
).properties(
    width=650,
    height=50
)

# scatter plot, modify opacity based on selection
countries = alt.Chart(data).mark_circle().encode(
    x='Population:Q',
    y='GDP:Q',
    tooltip='Country:N',
    opacity=alt.condition(brush, alt.value(0.75), alt.value(0.05))
).properties(
    width=650,
    height=400
)

alt.vconcat(birth_rate, countries).properties(spacing=5)
```
