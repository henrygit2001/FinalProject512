```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
data = pd.read_csv('world-data-2023.csv')
#keep = data.loc[:0, 'Agricultural Land( %)':'Unemployment rate'].columns
keep = ['Density\r\n(P/Km2)','Population','GDP','Agricultural Land( %)','Forested Area (%)']
data = data.dropna(axis=0,subset = keep)
data.head(10)
```

```python
data.rename(columns={'Density\r\n(P/Km2)': 'PopulationDensity'}, inplace=True)
data.rename(columns={'Forested Area (%)': 'ForestedArea'}, inplace=True)
```

```python
data.head()
```

```python
print(data.shape)
```

```python
print(data.columns.tolist())
```

```python
data['Population'] = data['Population'].str.replace(',', '').astype(float)

data['PopulationDensity'] = data['PopulationDensity'].str.replace(',', '').astype(float)
# data['Birth Rate'] = data['Birth Rate'].str.replace(',', '').astype(float)
data['GDP'] = data['GDP'].str.replace(',', '').str.replace('$', '').astype(float)
data['Agricultural Land( %)'] = data['Agricultural Land( %)'].str.replace(',', '').str.replace('%', '').astype(float)
data['ForestedArea'] = data['ForestedArea'].str.replace(',', '').str.replace('%', '').astype(float)
data['Armed Forces size'] = data['Armed Forces size'].str.replace(',', '').str.replace('%', '').astype(float)
```

```python
data['PopulationDensity']
```

```python

brush = alt.selection_interval(
    encodings=['x'] # limit selection to x-axis (year) values
)

brush2 = alt.selection_interval(
    encodings=['x'] # limit selection to x-axis (year) values
)

# dynamic query histogram
birth_rate = alt.Chart(data).mark_bar().add_params(
    brush
).encode(
    alt.X('ForestedArea:Q', title='Forested Area (%)'),
    alt.Y('count():Q', title=None)
).properties(
    width=650,
    height=50
)

agricultural_land = alt.Chart(data).mark_bar().add_params(
    brush2
).encode(
    alt.X('Agricultural Land( %):Q', title='Agricultural Land (%)'),
    alt.Y('count():Q', title=None)
).properties(
    width=650,
    height=50
)

# scatter plot, modify opacity based on selection
countries = alt.Chart(data).mark_circle().encode(
    x='PopulationDensity:Q',
    y='Birth Rate:Q',
    tooltip='Country:N',
    opacity=alt.condition(brush, alt.value(0.75), alt.value(0.20)),
    fillOpacity=alt.condition(brush2, alt.value(0.75), alt.value(0.20))
).properties(
    width=650,
    height=400
)

alt.vconcat(countries, birth_rate, agricultural_land).properties(spacing=10)
```

```python
gap = pd.read_csv('ddf--datapoints--alternative_gdp_per_capita_ppp_wb--by--geo--time.csv')
#gap = gap.melt(id_vars=['country'], var_name="Year", value_name='GDPpc')

gap2 = pd.read_csv('ddf--datapoints--at_least_basic_sanitation_overall_access_percent--by--geo--time.csv')
#gap2 = gap2.melt(id_vars=['country'], var_name="Year", value_name='Sanitation')

#gap3 = pd.read_csv('ddf--datapoints--total_population_with_projections--by--geo--time.csv')
#gap3.columns = gap3.columns.str.replace('name', 'country')
#gap3.columns = gap3.columns.str.replace('time', 'Year')

gap3 = pd.read_csv('pop_number.csv')

region = pd.read_csv('Data Geographies - v2 - by Gapminder - list-of-countries-etc.csv')
#region.columns = region.columns.str.replace('name', 'country')

gapcombo = pd.merge(gap, gap2, on=['geo','time'])#,how='outer')
# gapcombo['Year'] = gapcombo['Year'].astype('int64')
gapcombo = pd.merge(gap3, gapcombo, on=['geo','time'])#,how='outer')
gapcombo = pd.merge(region, gapcombo, on=['geo','name'])



# Add extra columns for Tooltip
gapcombo['at_least_basic_sanitation_overall_access_percentage'] = (gapcombo['at_least_basic_sanitation_overall_access_percent']).round(2).astype(str) + '%'
gapcombo['alternative_gdp_per_capita_ppp_wb_new'] = '$' + (gapcombo['alternative_gdp_per_capita_ppp_wb']).round(2).astype(str)
gapcombo['population_new'] = gapcombo['Population'].astype(int).apply('{:,}'.format)



# Make the larger circles appear below the smaller circles
gapcombo = gapcombo.sort_values(by='Population', ascending=False)




#gapcombo = gap.loc[(gap['country'].str.contains("USA"))]
gapcombo.head(5)
#gap2.info()





```

```python
# For the alt.Color legend:
regions = ['North America', 'Latin America & Caribbean', 'Europe & Central Asia', 'East Asia & Pacific', 'South Asia', 'Middle East & North Africa', 'Sub-Saharan Africa']
colorings = ['lightgreen', 'green', 'lightblue', 'blue', 'brown', 'orange', 'red']



slider = alt.binding_range(min=2000, max=2022, step=1, name='Current Year: ')

selector = alt.selection_point(
    name="SelectorName",
    fields=['time'],
    bind=slider,
    value=[{'time': 2000}]
)

search_input = alt.param(
    value = '',
    bind = alt.binding(
        input='search',
        placeholder="Country Name",
        name='Search For A Country: ',
    )
)

tracing_lines = alt.Chart(gapcombo).transform_calculate(
    trace_x='datum.alternative_gdp_per_capita_ppp_wb',
    trace_y='datum.at_least_basic_sanitation_overall_access_percent'
).mark_line(color='red', strokeDash=[3,3], opacity=0.25).encode(
    x='trace_x:Q',
    y='trace_y:Q',
    detail='name:N'
).transform_filter(
    alt.expr.test(alt.expr.regexp(search_input, 'i'), alt.datum.name) & selector &
    (alt.datum.time <= selector.time)
)


graph = alt.Chart(gapcombo).mark_circle(stroke='black', strokeWidth=1.5, strokeOpacity=0.75).encode(
    alt.Size('Population:Q',
              scale=alt.Scale(range=[100, 5000], bins=[0, 25000000, 50000000, 100000000, 200000000, 300000000, 1000000000, 2000000000]),
              title='Size of Population', 
              legend=alt.Legend(
                orient='none',
                legendX=764,
                legendY=180,
                titleAlign='left',
                titlePadding=8,
                titleFontSize=11.5,
                labelFontSize=10.5
              )
              ),
    alt.Color('World bank region:N',
                scale=alt.Scale(domain=regions, range=colorings), 
                title='World Regions',
                legend=alt.Legend(
                    orient='none',
                    legendX=764,
                    legendY=10,
                    titleAlign='left',
                    titlePadding=5,
                    titleFontSize=11.5,
                    labelFontSize=10.5
              )
              ), # many other region choices
    alt.X('alternative_gdp_per_capita_ppp_wb:Q',
            scale=alt.Scale(type='log',domain=[400,150000]),
            title='GDP per capita, PPP in international dollar (log scale)'
          ),
    alt.Y('at_least_basic_sanitation_overall_access_percent:Q',
            scale=alt.Scale(domain=[0, 104]),
            title='Percentage of Population with access to Basic Sanitation (%)'
          ),
    # tooltip=[
    #     'name:N',
    #     'alternative_gdp_per_capita_ppp_wb:Q',
    #     'at_least_basic_sanitation_overall_access_percent:Q'
    #     ],
    tooltip=[
        alt.Tooltip('name:N', title='Country Name:'),
        alt.Tooltip('population_new:N', title='Population: '),
        alt.Tooltip('World bank region:N', title='World Region: '),
        alt.Tooltip('alternative_gdp_per_capita_ppp_wb_new:N', title='GDP per capita, PPP (international dollar):'),
        alt.Tooltip('at_least_basic_sanitation_overall_access_percentage:N', title='Share of Population (%) with access to Basic Sanitation:')
        ],
    #opacity=alt.condition(brush, alt.value(0.75), alt.value(0.20)),
    #fillOpacity=alt.condition(brush2, alt.value(0.75), alt.value(0.20))
    opacity=alt.condition(
        alt.expr.test(alt.expr.regexp(search_input, 'i'), alt.datum.name),
        alt.value(0.75),
        alt.value(0.05)
    ),
).add_params(
    selector,
    search_input
).transform_filter(
    selector
).properties(
    width=750,
    height=500,
    title=alt.TitleParams(
        text="How have GDPs and Basic Sanitation accessibility changed over time?",
        subtitle=["(Circles represent individual countries. When varying the year, some circles may appear/disappear due to insufficient data.)", "", "This interactive visualization has tooltips, panning, and zooming! Double-click to go back to the original state!"],
        fontSize=18.5,
        subtitleFontSize=13,
        dx=72
    )
).configure_axis(
    titleFontSize=11.5,
    labelFontSize=10.5
).interactive()


graph





# layer = graph
# if search_input != '':
#     layer += tracing_lines

# layer.configure_axis(
#     titleFontSize=11.5,
#     labelFontSize=10.5
# )


# # st.altair_chart(layer, use_container_width=True)
# # layer.show()

# layer
```
